```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'literature_values.csv'
data = pd.read_csv(file_path)


# Define markers and colors for different Cu content
markers = {10: '>', 15: 'd', 20: '^', 25: '*'}
colors = {10: 'red', 15: 'green', 20: 'blue', 25: 'darkkhaki'}

# Create a new figure
plt.figure(figsize=(6, 4))

# Plot the data
for cu_content in data['Cu content'].unique():
    subset = data[data['Cu content'] == cu_content]
    plt.plot(subset['Cooling rate'], subset['Nucleation rate'], marker=markers[cu_content], color=colors[cu_content], label=f'Al-{cu_content} wt% Cu (Literature)', linestyle='dashed', markersize=10, linewidth=2)

# Add labels and title
plt.xlabel('Cooling rate (K s$^{-1}$)')
plt.ylabel('Crystal formation rate (crystal mm$^{-3}$ s$^{-1}$)')
plt.title('Crystal Formation Rate vs Cooling Rate')

# Remove box around legend
plt.legend(title='Cu content', frameon=False, loc='lower right')


# Show the plot
plt.show()
```
